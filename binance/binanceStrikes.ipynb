```python
import pandas as pd
```

```python
def GetCallVolumeByStrike(dateStr):
    #2023-08-18
    df = pd.read_csv(f"../../marketData/BTCUSDT-EOHSummary-{dateStr}.csv")
    df = df[df.volume_contracts > 0]
    df = df.drop(columns=['gamma', 'vega', 'theta', 'delta', 'best_buy_iv', 'best_sell_iv', 'open', 'close', 'high', 'low', 'mark_price', 'mark_iv', 'openinterest_usdt', 'best_bid_price', 'best_ask_price', 'best_bid_qty', 'best_ask_qty', 'volume_usdt', 'openinterest_contracts', 'underlying', 'symbol'])
    summed = df.groupby(['date', 'strike', 'type'], as_index = False)['volume_contracts'].sum()
    return summed[summed.type=="C"].drop(columns=['type', 'date'])
```

```python
dfs = []
for date in ['14', '15', '16', '17', '18']:
    dateStr = f"2023-08-{date}"
    dfs.append(GetCallVolumeByStrike(dateStr))
```
