```python
import numpy as np
import pandas as pd
```

## Fixed Rate

```python
df = pd.DataFrame(np.random.normal(100, 20, 1000) / 100)
df.columns = ["rate"]
df = df.sort_values(by="rate").reset_index(drop=True)
```

```python
df["rate^3"] = np.power(df["rate"], 3)
```

```python
df.plot.hist(bins=50, alpha=0.5)
```

## Normally Distributed Rate

```python
data = dict()
for i in range(5):
    data[f"rate_{i}"] = (np.random.normal(100, 20, 1000) / 100)
```

```python
df = pd.DataFrame(data)
```

```python
df['final'] = df['rate_0'] * df['rate_1'] * df['rate_2'] * df['rate_3'] * df['rate_4']
```

```python
df[['rate_0', 'final']].plot.hist(bins=50, alpha=0.5)
```
